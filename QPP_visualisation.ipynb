```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

df_reported = pd.read_csv("./QPP_scores_new/output/reported_values.csv")
df_reported
```

```python
kid_friend_df = df_reported.loc[df_reported["dataset"]=="kid-friend-en"]
kid_friend_df_corrP = kid_friend_df.loc[kid_friend_df["metric"]=="corrPearson"]

# a4_dims = (11.7, 8.27)
# fig, ax = plt.subplots(figsize=a4_dims)
g = sns.catplot(x=kid_friend_df_corrP["QPP"], y=kid_friend_df_corrP["bootstrap_avg"], hue=kid_friend_df_corrP["rel_type"], kind="bar", height=3, aspect=3)
# g.despine(left=True)
g.set_xticklabels(rotation=60)
g.fig.suptitle("Change in QPP peformance (Pearson's correlation) across relevance definitions.")
g.set_axis_labels("QPP", "Pearson's correlation")
g.legend.set_title("Relevance definition")
# sns.move_legend(g, "upper right", bbox_to_anchor=(0.5, -0.2), ncol=4)
sns.move_legend(g, "upper right", bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
import pandas as pd
import dataframe_image as dfi

df = pd.read_csv("./visualisation_data/kid-friend-en_qpp_scores.csv").fillna(0)
df_cols = df.columns.to_list()
corrP_cols = [col for col in df.columns if col.endswith("corrP")]
corrK_cols = [col for col in df.columns if col.endswith("corrK")]
corrS_cols = [col for col in df.columns if col.endswith("corrS")]
smare_cols = [col for col in df.columns if col.endswith("smare")]
georisk_cols = [col for col in df.columns if col.endswith("georisk")]
```

```python
df_corrP = df[["QPP"]+corrP_cols]
df_corrP.columns = [col.split(":")[0] for col in df_corrP.columns]
df_corrP = df_corrP.rename(columns={"QPP":"QPP:corrP"})
df_styled_corrP = df_corrP.style.format(precision=3).background_gradient(cmap='Greens',axis=1)
df_styled_corrP.export_png("./visualisation_data/rel_def_corrP.png", table_conversion="matplotlib")
df_styled_corrP
```

```python
df_smare = df[["QPP"]+smare_cols]
df_smare.columns = [col.split(":")[0] for col in df_smare.columns]
df_smare = df_smare.rename(columns={"QPP":"QPP:sMARE"})
df_styled_smare = df_smare.style.format(precision=3).background_gradient(cmap='Blues',axis=1)
df_styled_smare.export_png("./visualisation_data/rel_def_smare.png", table_conversion="matplotlib")
df_styled_smare
```

```python
df_georisk = df[["QPP"]+georisk_cols]
df_georisk.columns = [col.split(":")[0] for col in df_georisk.columns]
df_georisk = df_georisk.rename(columns={"QPP":"QPP:GeoRisk"})
df_styled_georisk = df_georisk.style.format(precision=3).background_gradient(cmap='Purples',axis=1)
df_styled_georisk.export_png("./visualisation_data/rel_def_georisk.png", table_conversion="matplotlib")
df_styled_georisk
```
